```python
from selenium import webdriver
import selenium.webdriver.support.ui as ui
import time

results = []
browser = webdriver.Chrome(executable_path='./chromedriver')
wait = ui.WebDriverWait(browser,10)
# implicitly_wait隱性等待5秒，等網頁載入完成才執行下一步
browser.implicitly_wait(5)
browser.get("https://hahow.in/courses")
# 取得資料
html_source = browser.page_source

max_page = 27
for page_number in range(1,max_page+1,1):
    soup = BeautifulSoup(html_source,"html.parser")
    content = soup.select('h4.title')
    for ele in content:
        results.append(ele.get_text())
    if page_number == max_page:
        break
    else:
        button = browser.find_element_by_xpath("//li[@title='下一頁']/a")
        browser.execute_script("arguments[0].click();",button)
    time.sleep(3)

print(results)
```
