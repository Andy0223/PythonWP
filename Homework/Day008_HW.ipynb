```python
import requests
from bs4 import BeautifulSoup
r1 = requests.get("https://www.dcard.tw/f")

headers = {'user-agent':'my-app/0.01'}
r2 = requests.get("https://www.zhihu.com/explore",headers=headers)

response = r1.text
response1 = r2.text

print(type(response))

print(response)
```

```python
#印出 Dcard 的 html 內容（已經整理過） 
soup = BeautifulSoup(response)
print(soup.prettify())
```

```python
#印出標籤和內部屬性
print(soup.title)
print(soup.span)
print(soup.find(id='titile-id'))
```

```python
#知乎
soup1 = BeautifulSoup(response1)
print(soup1.prettify())
```
